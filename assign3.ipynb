```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('admission.csv')
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(data['GRE Score'], data['CGPA'], c='orange', alpha=1,)
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Scatter Plot')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(data['University Rating'], data['CGPA'], color= 'red')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.title('Bar Plot')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x=data['University Rating'], y=data['TOEFL Score'])
plt.title('Box Plot')
plt.show()
```

```python
sns.boxplot(x = "GRE Score", data = data, width = 0.8, showmeans = True)
```

```python
# Pie Chart
plt.figure(figsize=(6,6))
research = data['Research'].value_counts()
plt.pie(research, labels=['Research (1)', 'No Research (0)'], autopct='%1.1f%%', colors=['orange', 'blue',])
plt.title('Pie Chart')
plt.show()
```

```python
# line chart
plt.plot(data["Serial No."], data["CGPA"])
plt.xlabel('Serial No.')
plt.ylabel('CGPA')
plt.title('Line Chart')
plt.show()
```
